```python
import matplotlib.pyplot as plt

# Ajuste lineal: canal = a * energía + b
a = 2.06
b = -27.94

# Energías conocidas de los picos de Eu-152 (en keV)
energias_keV = [
    121.7817,
    244.6974,
    344.2785,
    411.1165,
    443.9606,
    778.9045,
    964.057,
    1085.837,
    1299.142,
    1408.013
]

def leer_spe_ascii(ruta_archivo):
    espectro = []
    dentro_de_datos = False
    with open(ruta_archivo, 'r') as f:
        for linea in f:
            linea = linea.strip()
            if linea.startswith('$DATA:'):
                dentro_de_datos = True
                inicio_final = next(f).strip()
                inicio, final = map(int, inicio_final.split())
                continue
            if dentro_de_datos:
                if linea.startswith('$'):
                    break
                if linea != '':
                    espectro.append(int(linea))
    return espectro

def energia_a_canal(energia):
    return a * energia + b

def graficar_espectro_con_picos(espectro):
    canales = list(range(len(espectro)))

    fig, ax = plt.subplots(figsize=(6, 5))

    ax.plot(canales, espectro, drawstyle='steps-mid')
    ax.set_xlabel('Canal')
    ax.set_ylabel('Cuentas')
    ax.grid(alpha=0.3)
    ax.set_ylim(0, max(espectro) * 1.1)

    for i, energia in enumerate(energias_keV):
        canal = int(energia_a_canal(energia))
        if 0 <= canal < len(espectro):
            cuenta = espectro[canal]

            if i == 0:
                # Flecha horizontal hacia la derecha
                ax.annotate(
                    f'{energia:.1f} keV',
                    xy=(canal, cuenta),
                    xytext=(20, 0),
                    textcoords='offset points',
                    ha='left',
                    va='center',
                    fontsize=8,
                    rotation=0,
                    arrowprops=dict(arrowstyle='->', lw=0.8)
                )
            else:
                # Flecha diagonal hacia arriba-derecha
                ax.annotate(
                    f'{energia:.1f} keV',
                    xy=(canal, cuenta),
                    xytext=(5, 15),
                    textcoords='offset points',
                    ha='left',
                    fontsize=8,
                    rotation=45,
                    arrowprops=dict(arrowstyle='->', lw=0.8)
                )

    # Eje superior: Energía
    ax2 = ax.secondary_xaxis('top', functions=(lambda c: (c - b) / a, energia_a_canal))
    ax2.set_xlabel('Energía (keV)')

    plt.tight_layout()
    plt.show()

# Leer y graficar
ruta = 'calibracion/eu152.6min.Spe'
espectro = leer_spe_ascii(ruta)
graficar_espectro_con_picos(espectro)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Datos
E = np.array([121.7817, 244.6974, 344.2787, 443.9606,
              1408.013, 1085.837, 511, 1274.537])

Eficiencia = np.array([
    0.004855477961117568, 0.002827151466973118, 0.0022048630494630886,
    0.0017913303346990099, 0.0006317650057242007, 0.0009753866688973304,
    0.0015972763038564228, 0.0007138220000563729
])

# Errores: 5% de cada eficiencia
err_eficiencia = 0.05 * Eficiencia

# Modelo de ajuste: eficiencia = a * E^b
def modelo(E, a, b):
    return a * E**b

# Ajuste con curve_fit
parametros, _ = curve_fit(modelo, E, Eficiencia)
a, b = parametros

print(f"Parámetro a: {a:.6e}")
print(f"Parámetro b: {b:.6f}")

# Generar puntos para la curva ajustada
E_fit = np.linspace(min(E), max(E), 500)
Eficiencia_fit = modelo(E_fit, a, b)

# Graficar
plt.errorbar(E, Eficiencia, yerr=err_eficiencia, fmt='o', color='blue', capsize=5, label='Datos experimentales')
plt.plot(E_fit, Eficiencia_fit, color='red', label=f'Ajuste: $\\epsilon = {a:.3f} \\cdot E^{{{b:.3f}}}$')
plt.xlabel('Energía (keV)')
plt.ylabel('Eficiencia')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
